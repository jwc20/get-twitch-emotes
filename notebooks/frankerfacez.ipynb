```python
import requests
import json
```

```python
channel_name = "sodapoppin" 
response = requests.get(f"https://api.frankerfacez.com/v1/room/{channel_name}")

if response.status_code == 200:
    data = response.json()
    emotes = data
    
```

```python
print(emotes)
```

```python
with open('../data/frankerfacez_emotes.json', 'w') as f:
    json.dump(emotes, f, indent=4)
```
