```python
# https://7tv.app/emotes

import requests
import json
```

```python
base_url = "https://7tv.io/v3"
endpoint = "/emotes"

response = requests.get(base_url + endpoint)

# https://7tv.app/emotes/60ae7316f7c927fad14e6ca2
catJAM_id = "60ae7316f7c927fad14e6ca2"
catJAM_response = requests.get(base_url + endpoint + "/" + catJAM_id)
```

```python
if response.status_code == 200:
    emotes = response.json()
    print(emotes)
else:
    print(f"Failed to fetch emotes. Status code: {response.status_code}")
```
